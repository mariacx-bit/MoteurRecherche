```python
def tableau_avant_apres(df, terme, annee_coupe):
    date_cut = pd.Timestamp(f"{annee_coupe}-01-01")
    total_before = compter_occurrences(df[df['date'] < date_cut], terme)
    total_after  = compter_occurrences(df[df['date'] >= date_cut], terme)

    return pd.DataFrame({
        'periode': ['avant', 'apres'],
        'occurrences': [total_before, total_after],
    })
```

```python
def compter_occurrences(df_part, terme):
    terme = terme.lower().strip()
    return df_part['text'].str.lower().str.count(terme).sum()
```
